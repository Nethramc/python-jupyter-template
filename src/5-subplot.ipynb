### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
import matplotlib.pyplot as plt
import numpy as np

# Data for plots
x = np.arange(1, 6)
y = x * 2
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
colors = ['red', 'blue', 'green', 'purple']
pie_values = [30, 20, 25, 25]
pie_labels = ['Slice 1', 'Slice 2', 'Slice 3', 'Slice 4']

# Create a 2x2 subplot layout
fig, axes = plt.subplots(2, 2, figsize=(10, 10))

# Line plot
axes[0, 0].plot(x, y, marker='o', linestyle='-', color='b')
axes[0, 0].set_title("Line Chart")
axes[0, 0].set_xlabel("X values")
axes[0, 0].set_ylabel("Y values")
axes[0, 0].grid(True, linestyle='--', linewidth=0.6)

# Bar chart
axes[0, 1].bar(categories, values, color=colors)
axes[0, 1].set_title("Bar Chart")
axes[0, 1].set_xlabel("Categories")
axes[0, 1].set_ylabel("Values")
axes[0, 1].grid(axis='y', linestyle='--', linewidth=0.6, alpha=0.7)

# Scatter plot
axes[1, 0].scatter(x, y, color='blue', label='Data Points')
for i in range(len(x)):
    axes[1, 0].annotate(f'({x[i]}, {y[i]})', (x[i], y[i]), textcoords="offset points", xytext=(5,5), ha='center')
axes[1, 0].set_title("Scatter Plot")
axes[1, 0].set_xlabel("X values")
axes[1, 0].set_ylabel("Y values")
axes[1, 0].grid(True, linestyle='--', linewidth=0.6)
axes[1, 0].legend()

# Pie chart
axes[1, 1].pie(pie_values, labels=pie_labels, autopct='%1.1f%%', colors=colors, startangle=140)
axes[1, 1].set_title("Pie Chart")

# Adjust layout and display the plot
plt.tight_layout()
plt.show()
```
