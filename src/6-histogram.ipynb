### Exercise 6: Histogram Customization

```python
# TASK: Create a histogram for the following data: data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5].
# Customize the histogram with a title, labels for the x-axis, and a specific color for the bars.
import matplotlib.pyplot as plt

# Data for histogram
data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5]

# Create the histogram
plt.figure(figsize=(8, 5))
plt.hist(data, bins=5, color='teal', edgecolor='black', alpha=0.7)

# Customize the plot
plt.title("Histogram of Data", fontsize=14, fontweight='bold')
plt.xlabel("Data Values", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.grid(axis='y', linestyle='--', linewidth=0.6, alpha=0.7)

# Display the plot
plt.show()
```
